## Menentukan Pertanyaan Bisnis

- Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?
- Bagaimana tren penggunaan sepeda berdasarkan jam dalam sehari?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Load dataset
day_df = pd.read_csv('data/day.csv')
hour_df = pd.read_csv('data/hour.csv')

day_df.head(), hour_df.head()
```

**Insight:**

- Dataset berisi informasi penyewaan sepeda harian dan per jam.
- Memiliki variabel yang dapat dianalisis lebih lanjut.

### Assessing Data

```python
day_df.info()
hour_df.info()
```

### Cleaning Data

```python
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
day_df['season'] = day_df['season'].map(season_mapping)
```

**Insight:**
- Variabel `season` telah dikonversi ke bentuk yang lebih mudah dipahami.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
day_df.describe(), hour_df.describe()
```

**Insight:**
- Rentang nilai variabel terlihat wajar dan tidak ada outlier ekstrem.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='season', y='cnt', data=day_df, hue='season', palette='coolwarm', errorbar=None, legend=False)  
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='hr', y='cnt', data=hour_df, marker='o', color='b')
plt.title('Tren Penyewaan Sepeda Berdasarkan Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

**Insight:**

- Musim gugur memiliki tingkat penyewaan tertinggi.
- Penyewaan sepeda lebih rendah pada musim semi.
- Penyewaan sepeda meningkat pada pagi dan sore hari, menunjukkan pola perjalanan kerja/sekolah.

## Analisis Lanjutan (Opsional)

## Conclusion

- Musim sangat mempengaruhi jumlah penyewaan sepeda. Musim gugur memiliki tingkat penyewaan tertinggi, sedangkan musim semi lebih rendah.
- Penyewaan sepeda lebih tinggi pada pagi dan sore hari, menunjukkan pola perjalanan kerja atau sekolah.
